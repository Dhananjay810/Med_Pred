```python
import pickle
import pandas as pd

# Load model
pipe_lr = pickle.load(open('pipe_lr.pkl','rb'))

# Create input with all zeros (377 features)
input_data = pd.DataFrame([[1]*377])

# 🔥 IMPORTANT: set column names (must match training)
input_data.columns = pipe_lr.feature_names_in_

# Predict
prediction = pipe_lr.predict(input_data)

print("Prediction:", prediction[0])
```
